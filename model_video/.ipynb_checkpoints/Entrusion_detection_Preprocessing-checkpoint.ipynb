```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from category_encoders import OneHotEncoder
import datetime

# Load your dataset
df = pd.read_csv(r"C:\Users\LENOVO\OneDrive\Desktop\MAJOR-PROJECT\enhanced_intrusion_detection_dataset.csv")

# Optional: Drop columns that are not useful for model learning
df = df.drop(columns=["session_id", "source_ip", "destination_ip"])

# Convert timestamp to datetime object
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Create new features from timestamp if desired
df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['weekday'] = df['timestamp'].dt.weekday

# Drop raw timestamp if not needed
df = df.drop(columns=['timestamp'])

# Identify categorical columns
categorical_cols = ['protocol_type', 'encryption_used', 'browser_type', 'user_agent', 'flag']

# One-hot encode categorical columns
encoder = OneHotEncoder(cols=categorical_cols, use_cat_names=True)
df_encoded = encoder.fit_transform(df)

# Separate features and labels
X = df_encoded.drop(columns=["attack_detected"])
y = df_encoded["attack_detected"]

# Scale numeric features
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Optional: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Check processed data shape
print("Processed Features Shape:", X_scaled.shape)
print("Target Label Distribution:\n", y.value_counts(normalize=True))
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE

# 1. Load your dataset
df = pd.read_csv("C:/Users/LENOVO/OneDrive/Desktop/MAJOR-PROJECT/enhanced_intrusion_detection_dataset.csv")

# 2. Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 3. Drop 'session_id' (not useful for modeling)
df.drop(['session_id'], axis=1, inplace=True)

# 4. Label encode categorical columns
categorical_cols = ['source_ip', 'destination_ip', 'protocol_type', 'encryption_used', 
                    'browser_type', 'user_agent', 'flag']
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# 5. Create time-based features and drop original timestamp
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df.drop('timestamp', axis=1, inplace=True)  # Drop timestamp as it's non-numeric

# 6. Separate features and target
X = df.drop('attack_detected', axis=1)
y = df['attack_detected']

# 7. Apply SMOTE to handle class imbalance
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

# 8. Scale numeric features
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_resampled), columns=X.columns)

# Final output
print("Preprocessing complete.")
print("X_scaled shape:", X_scaled.shape)
print("y_resampled distribution:\n", pd.Series(y_resampled).value_counts())
```

```python
import os

# 1. Define the folder where you want to save the processed dataset
output_folder = "C:/Users/LENOVO/OneDrive/Desktop/MAJOR-PROJECT/Processed_Data"

# 2. Create the folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# 3. Combine features and target for saving
df_processed = X_scaled.copy()
df_processed['attack_detected'] = y_resampled.values

# 4. Define the output file path
output_file = os.path.join(output_folder, "preprocessed_intrusion_data.csv")

# 5. Save to CSV
df_processed.to_csv(output_file, index=False)

print("✅ Preprocessed dataset saved at:", output_file)
```
